# **Data Cleaning**

```python
print("rows,cols:",df.shape)
print(df.columns.tolist())
print("Num duplicates:",df.duplicated().sum())
print("Missing Values:")
print(df.isna().sum().sort_values(ascending=False).head(20))
```

```python
columns_to_drop = [
    'Customer Email', 'Customer Fname', 'Customer Lname', 'Customer Password',
    'Customer Street', 'Customer Zipcode', 'Product Description', 'Product Image',
    'Order Zipcode', 'Latitude', 'Longitude', 'Product Status',
    'Category Id', 'Department Id', 'Product Card Id', 'Product Category Id',
    'Order Item Cardprod Id', 'Order Item Id'
]

df = df.drop(columns=columns_to_drop, errors='ignore')
df=df[df['Delivery Status']!='Shipping canceled']
df=df[df['Delivery Status']!='Returned']
df=df[df['Delivery Status']!='Cancelled']
df=df[df['Delivery Status']!='Processing']
df=df[df['Delivery Status']!='Payment failure']
df=df[df['Delivery Status']!='Order Returned']
df=df[df['Delivery Status']!='Order Refunded']
for c in ['order date (DateOrders)','shipping date (DateOrders)']:
    df[c]=pd.to_datetime(df[c], errors='coerce', dayfirst=False)
print("DataFrame shape after dropping columns:", df.shape)
print("Missing Values:")
print(df.isna().sum().sort_values(ascending=False).head(5))
```
